## Clean the Data

```python
vehicles.drop(['region_url','url','posting_date','lat','long','image_url','county','VIN','description'],axis=1,inplace=True)
vehicles.dropna(thresh=13,inplace=True)
vehicles
```

```python
bigprice = vehicles[vehicles['price']>=350000].index
vehicles.drop(bigprice,inplace=True)
vehicles.groupby(['type'])['price'].max()

lowprice = vehicles[vehicles['price']<=500].index
vehicles.drop(lowprice,inplace=True)
vehicles.groupby(['type'])['price'].min()

lowodometer = vehicles[vehicles['odometer']<200].index
vehicles.drop(lowodometer,inplace=True)
vehicles.groupby(['type'])['odometer'].min()

highodometer = vehicles[vehicles['odometer']>250000].index
vehicles.drop(highodometer,inplace=True)
vehicles.groupby(['type'])['odometer'].max()

dropoldcars = vehicles[vehicles['year']<2000].index
vehicles.drop(dropoldcars,inplace=True)
vehicles.groupby(['manufacturer'])['year'].min()

newcars = vehicles[vehicles['condition']=='new'].index
vehicles.drop(newcars,inplace=True)
vehicles.groupby(['condition'])['odometer'].min()

other = vehicles[vehicles['type']=='other'].index
vehicles.drop(other,inplace=True)
vehicles.groupby(['type'])['price'].min()

bus = vehicles[vehicles['type']=='bus'].index
vehicles.drop(bus,inplace=True)
vehicles.groupby(['type'])['price'].min()
```
